```python
import turtle

screen = turtle.Screen()
screen.bgcolor("lightblue")

pen = turtle.Turtle()

#Vẽ mặt trời
pen.penup()
pen.goto(200, 450)
pen.pendown()

for _ in range(8):
        pen.forward(100)
        pen.backward(100)
        pen.left(45)

pen.penup()
pen.backward(60)
pen.right(90)
pen.pendown()

pen.fillcolor ("yellow")
pen.begin_fill()
pen.circle (60)
pen.end_fill()



#Vẽ nhà
pen.penup()
pen.goto(-320, -50)
pen.pendown()
pen.setheading(0)

#tam giác
pen.fillcolor ("pink")
pen.begin_fill()
pen.left(45)
pen.forward(200)
pen.right(90)
pen.forward(200)
pen.right(135)
pen.forward(283)
pen.end_fill()

#chữ nhật
pen.fillcolor ("blue")
pen.begin_fill()
pen.left(90)
pen.forward(320)
pen.left(90)
pen.forward(283)
pen.left(90)
pen.forward(320)
pen.end_fill()

#bình hành
pen.fillcolor ("orange")
pen.begin_fill()
pen.right(65)
pen.forward(300)
pen.left(110)
pen.forward(200)
pen.left(70)
pen.forward(300)
pen.end_fill()

# pen.right(175)

pen.penup()
pen.goto(-37, -370)
pen.pendown()

pen.fillcolor ("yellow")
pen.begin_fill()
pen.setheading(0)
pen.left(25)
pen.forward(300)
pen.left(65)
pen.forward(320)
pen.left(115)
pen.forward(300)
pen.end_fill()

#cửa nhà
pen.penup()
pen.goto(-204, -370)
pen.pendown()

pen.fillcolor ("yellow")
pen.begin_fill()
pen.setheading(0)

pen.left(90)
pen.forward(100)
pen.right(90)
pen.forward(50)
pen.right(90)
pen.forward(100)
pen.end_fill()

#cửa sổ
pen.penup()
pen.goto(90, -197)
pen.pendown()

pen.fillcolor ("blue")
pen.begin_fill()
pen.setheading(0)

pen.left(25)
pen.forward(50)
pen.left(65)
pen.forward(50)
pen.left(115)
pen.forward(50)
pen.left(65)
pen.forward(50)
pen.end_fill()



#Cây

pen.penup()
pen.goto(400, -370)
pen.pendown()

pen.fillcolor ("brown")
pen.begin_fill()
pen.setheading(0)
pen.left(90)
pen.forward(100)
pen.left(90)
pen.forward(26)
pen.left(90)
pen.forward(100)
pen.left(90)
pen.forward(26)
pen.end_fill()

pen.penup()
pen.goto(387, -220)
pen.pendown()

pen.fillcolor ("green")
pen.begin_fill()
pen.right(45)
pen.forward(100)
pen.right(135)
pen.forward(141)
pen.right(135)
pen.forward(100)


pen.penup()
pen.goto(387, -150)
pen.pendown()
pen.setheading(0)
pen.right(45)
pen.forward(100)
pen.right(135)
pen.forward(141)
pen.right(135)
pen.forward(100)

pen.penup()
pen.goto(387, -80)
pen.pendown()
pen.setheading(0)
pen.right(45)
pen.forward(100)
pen.right(135)
pen.forward(141)
pen.right(135)
pen.forward(100)

pen.end_fill()

turtle.done()
```
