```python
from tkinter import *
from tkinter import ttk, messagebox
import json
import os

class KhachHang:
    def __init__(self, ten, ngaysinh, lan_dau, lan_2, lan_3):
        self.ten = ten
        self.ngaysinh = ngaysinh
        self.lan_dau = lan_dau
        self.lan_2 = lan_2
        self.lan_3 = lan_3

class QuanLyKhachHang:
    def __init__(self, master):
        self.master = master
        self.master.title("Quản Lý Khách Hàng")
        self.master.geometry('')
        self.master.resizable(False, False)


        self.khachhang = []

        self.tao_widgets()

    def tao_widgets(self):
        Label(self.master, text="Tên:").grid(row=0, column=0)
        Label(self.master, text="Ngày Sinh:").grid(row=1, column=0)
        Label(self.master, text="Lần 1:").grid(row=2, column=0)
        Label(self.master, text="Lần 2:").grid(row=3, column=0)
        Label(self.master, text="Lần 3:").grid(row=4, column=0)

        self.ten_var = StringVar()
        self.ngaysinh_var = StringVar()
        self.lan_dau_var = StringVar()
        self.lan_2_var = StringVar()
        self.lan_3_var = StringVar()

        Entry(self.master, textvariable=self.ten_var).grid(row=0, column=1)
        Entry(self.master, textvariable=self.ngaysinh_var).grid(row=1, column=1)
        Entry(self.master, textvariable=self.lan_dau_var).grid(row=2, column=1)
        Entry(self.master, textvariable=self.lan_2_var).grid(row=3, column=1)
        Entry(self.master, textvariable=self.lan_3_var).grid(row=4, column=1)

        Button(self.master, text="Thêm", command=self.them_khachhang).grid(row=5, column=0)
        Button(self.master, text="Tìm Kiếm", command=self.tim_kiem_khachhang).grid(row=5, column=1)
        Button(self.master, text="Cập Nhật", command=self.cap_nhat_khachhang).grid(row=6, column=0)
        Button(self.master, text="Xóa", command=self.xoa_khachhang).grid(row=6, column=1)
        Button(self.master, text="Save", command=self.luu_ra_json).grid(row=6, column=2)
        Button(self.master, text="Reset", command=self.reset_entries).grid(row=5, column=2)

        self.tree = ttk.Treeview(self.master, columns=('Tên', 'Ngày Sinh', 'Lần 1', 'Lần 2', 'Lần 3'))
        self.tree.grid(row=7, column=0, columnspan=5)
        
        # Đặt tiêu đề cho các cột trong treeview
        self.tree.heading('Tên', text='Tên')
        self.tree.heading('Ngày Sinh', text='Ngày Sinh')
        self.tree.heading('Lần 1', text='Lần 1')
        self.tree.heading('Lần 2', text='Lần 2')
        self.tree.heading('Lần 3', text='Lần 3')
        
        # Cài đặt chiều rộng cho cột treeview
        self.tree.column('#0', width=0, stretch=NO)
        self.tree.column('Tên', anchor=W, width=150)
        self.tree.column('Ngày Sinh', anchor=W, width=100)
        self.tree.column('Lần 1', anchor=W, width=150)
        self.tree.column('Lần 2', anchor=W, width=100)
        self.tree.column('Lần 3', anchor=W, width=100)

    def reset_entries(self):
        # Reset các khung nhập
        self.ten_var.set("")
        self.ngaysinh_var.set("")
        self.lan_dau_var.set("")
        self.lan_2_var.set("")
        self.lan_3_var.set("")

    def them_khachhang(self):
        ten = self.ten_var.get()
        ngaysinh = self.ngaysinh_var.get()
        lan_dau = self.lan_dau_var.get()
        lan_2 = self.lan_2_var.get()
        lan_3 = self.lan_3_var.get()

        khachhang_moi = KhachHang(ten, ngaysinh, lan_dau, lan_2, lan_3)
        self.khachhang.append(khachhang_moi)

        self.tree.insert('', 'end', values=(ten, ngaysinh, lan_dau, lan_2, lan_3))
        messagebox.showinfo("Thành Công", "Khách Hàng đã được thêm!")
        self.reset_entries()

    def tim_kiem_khachhang(self):
        ten_tim_kiem = self.ten_var.get()

        for khachhang in self.khachhang:
            if khachhang.ten == ten_tim_kiem:
                messagebox.showinfo("Kết Quả Tìm Kiếm", f"Khách Hàng Được Tìm Thấy: \n Tên: {khachhang.ten}, Ngày Sinh: {khachhang.ngaysinh}, Ngày Đến Lần Đầu: {khachhang.lan_dau}, Lần 2: {khachhang.lan_2}, Lần 3: {khachhang.lan_3}")
                return

        messagebox.showinfo("Kết Quả Tìm Kiếm", "Khách Hàng không được tìm thấy.")

    def cap_nhat_khachhang(self):
        selected_item = self.tree.selection()  # Lấy dòng đã chọn trong treeview
        if selected_item:  # Kiểm tra xem có dòng nào được chọn không
            item_chon = selected_item[0]
            khachhang_chon = self.tree.item(item_chon, 'values')
    
            ten_moi = self.ten_var.get() if self.ten_var.get() else khachhang_chon[0]
            ngaysinh_moi = self.ngaysinh_var.get() if self.ngaysinh_var.get() else khachhang_chon[1]
            lan_dau_moi = self.lan_dau_var.get() if self.lan_dau_var.get() else khachhang_chon[2]
            lan_2_moi = self.lan_2_var.get() if self.lan_2_var.get() else khachhang_chon[3]
            lan_3_moi = self.lan_3_var.get() if self.lan_3_var.get() else khachhang_chon[4]
    
            for khachhang in self.khachhang:
                if khachhang.ten == khachhang_chon[0]:
                    khachhang.ten = ten_moi  # Cập nhật tên nếu muốn thay đổi tên
                    khachhang.ngaysinh = ngaysinh_moi
                    khachhang.lan_dau = lan_dau_moi
                    khachhang.lan_2 = lan_2_moi
                    khachhang.lan_3 = lan_3_moi
                    
                    # Cập nhật treeview
                    self.tree.set(item_chon, 'Tên', ten_moi)  # Cập nhật tên trong treeview
                    self.tree.set(item_chon, 'Ngày Sinh', ngaysinh_moi)
                    self.tree.set(item_chon, 'Lần 1', lan_dau_moi)
                    self.tree.set(item_chon, 'Lần 2', lan_2_moi)
                    self.tree.set(item_chon, 'Lần 3', lan_3_moi)
                    
                    messagebox.showinfo("Cập Nhật Khách Hàng", f"Cập nhật thông tin cho {ten_moi} thành công!")
                    break
        else:
            messagebox.showinfo("Cập Nhật Khách Hàng", "Vui lòng chọn một khách hàng để cập nhật!")



    def xoa_khachhang(self):
        item_chon = self.tree.selection()[0]
        khachhang_chon = self.tree.item(item_chon, 'values')

        for i, khachhang in enumerate(self.khachhang):
            if khachhang.ten == khachhang_chon[0]:
                del self.khachhang[i]
                self.tree.delete(item_chon)
                messagebox.showinfo("Thành Công", "Khách Hàng đã được xóa!")
                return

        messagebox.showinfo("Lỗi Xóa", "Không thể xóa Khách Hàng!")

    def luu_ra_json(self):
        if not self.khachhang:
            messagebox.showinfo("Lỗi Lưu", "Không có dữ liệu để lưu!")
            return

        du_lieu = []
        for khachhang in self.khachhang:
            du_lieu.append({
                "Tên": khachhang.ten,
                "Ngày Sinh": khachhang.ngaysinh,
                "Lần 1": khachhang.lan_dau,
                "Lần 2": khachhang.lan_2,
                "Lần 3": khachhang.lan_3
            })

        ten_file = 'khachhang.json'
        with open(ten_file, 'w') as f:
            json.dump(du_lieu, f)
        messagebox.showinfo("Thành Công", f"Dữ liệu đã được lưu vào {os.path.abspath(ten_file)}")

window = Tk()
app = QuanLyKhachHang(master=window)
window.mainloop()
```
